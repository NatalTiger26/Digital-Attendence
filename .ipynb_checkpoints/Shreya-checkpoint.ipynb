```python
#STEP3: Calculate the Covariance Matrix or the Sigma
covariance_matrix = []
# for i in range(len(normalized_face_vector)):
#     covariance_matrix.append(np.cov(np.transpose(normalized_face_vector[i])))
    
covariance_matrix.append(np.cov(np.transpose(avg_face_vector)))

print("Covariance Matrix",covariance_matrix)
```

```python
for i in range(len(covariance_matrix)):
    #STEP4: Calculate Eigen Vectors
    eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix[i])
    #STEP5: Select the K best Eigen Faces, K < M
    k = 50
    k_eigen_vectors = eigen_vectors[0:k, :]
    print("K Eigen Vectors Shape",k_eigen_vectors.shape)
    #STEP6: Convert lower dimensionality K Eigen Vectors to Original Dimensionality
    eigen_faces = k_eigen_vectors.dot(np.transpose(normalized_face_vector))
    print("Eigen Faces",eigen_faces, "Eigen Faces Shape",eigen_faces.shape)
    # STEP7: Represent Each eigen face as combination of the K Eigen Vectors
    weights = np.transpose(normalized_face_vector).dot(np.transpose(eigen_faces))
    print("Weights",weights)
```

```python
def distance_test_new (name, number):
    filename = name + "\\" + name + "_" + str(number) + ".jpg"
    distance = []
    test_img = cv2.imread(filename)
    test_img = cv2.cvtColor(test_img, cv2.COLOR_RGB2GRAY)
#     plt.imshow(test_img, cmap='gray')
#     plt.show()
    
    # For the test image making good box for pic leads to result being correct but when it is chossen 
    # with background then we get bad results
    #test_img = test_img[40:190, 30:180]
    
    test_img= cv2.resize(test_img, (image_width, image_length))
#     plt.imshow(test_img, cmap='gray')
#     plt.show()
    test_img = test_img.reshape(total_pixels, 1)
    
    # Normalizint the test pic leads to rreally bad results idk why
#     test_img = test_img * 10000 / sum(test_img) 
    
    count = 0
    for i in avg_face_vector:
        dist = np.linalg.norm(i - test_img)
        distance.append([celebrity_list[count], dist])
        count += 1
    distance.sort(key = lambda x: x[1]) # don't even ask i googled how to sort a list using second element of a list 
                                         # and got this. I don't know lambda functions.
    print("Actual name: ", name)
    print("Prediction : ", distance[0][0])
    for i in range(len(distance)):
        if distance[i][0] == name:
            print("Rank of the actual name : ", i +1)
            print("Percentage difference in distance of actual vs prediction : ", 100 * (distance[i][1] - distance[0][1])/distance[1][1])
    print("\n")
```
